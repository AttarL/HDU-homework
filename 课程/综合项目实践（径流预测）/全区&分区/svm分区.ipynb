```python
import os
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import time

# è®¾ç½®æ°æ®ç®å½
clusteri = 4
data_dir = f'cluster_{clusteri}'
files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith('.csv')]

# åå§åç»æè®°å½
results = []
time1 = time.time()
sumfiler=0
sumfile=0
# éåæ¯ä¸ªæµåçæ°æ®æä»¶
for file in files:
    sumfile=sumfile+1
    # è¯»åæ°æ®
    data = pd.read_csv(file)
    
    # ç¹å¾åç®æ 
    features = ['Dayl', 'Prcp', 'Srad', 'Tmax',
                'Prcp_lag_1', 'Discharge_lag_1', 'Prcp_lag_2', 'Discharge_lag_2',
                'Discharge_lag_3', 'Discharge_lag_4']
    target = 'Discharge'
    
    # åå²è®­ç»éåæµè¯é
    train_data = data.iloc[:1463]  # å1463å¤©çæ°æ®ä½ä¸ºè®­ç»é
    test_data = data.iloc[1463:]  # å¶ä½çæ°æ®ä½ä¸ºæµè¯é
    
    # ç¡®ä¿è®­ç»éåæµè¯éä¸ä¸ºç©º
    if train_data.empty or test_data.empty:
        print(f"Warning: {file} çè®­ç»éææµè¯éä¸ºç©ºï¼è·³è¿è¯¥æµåã")
        continue
    
    # åç¦»ç¹å¾åç®æ 
    X_train = train_data[features]
    y_train = train_data[target]
    X_test = test_data[features]
    y_test = test_data[target]
    
    # æ åå
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # åå§åæ¯æåéæºæ¨¡å
    svm_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
       # å®ä¹ç½æ ¼æç´¢çåæ°ç©ºé´
    param_grid = {
            'C': [0.1, 1, 10, 100],
            'kernel': ['linear', 'rbf'],
            'gamma': [0.1, 1, 10, 100]
    }

    # ä½¿ç¨Grid Search CVè¿è¡è¶åæ°è°ä¼
    grid_search = GridSearchCV(svm_model, param_grid, scoring='r2', cv=5)
    grid_search.fit(X_train_scaled, y_train)
     # è®­ç»æ¨¡å
    svm_model.fit(X_train_scaled, y_train)
    
    # é¢æµ
    y_pred = svm_model.predict(X_test_scaled)
    
    # è®¡ç®MSEåRÂ²
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f'{file} - Mean Squared Error: {mse}, RÂ²: {r2}')
    sumfiler=sumfiler+r2
    # ä¿å­ç»æ
    results.append({'æµå': file, 'MSE': mse, 'RÂ²': r2})
    
    # ç»å¶é¢æµç»æä¸çå®ç»æçå¯¹æ¯
    plt.figure(figsize=(10, 5))
    plt.plot(y_test.values, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title(f'{file} - æµåé¢æµå¯¹æ¯')
    plt.legend()
    plt.savefig(f'{file}_prediction_comparison.png')  # ä¿å­å¾ç
    plt.show()

time2 = time.time()

# è¾åºæ»æ¶é´
print(f'Total time taken: {time2 - time1} seconds')


```

```python
import pandas as pd
import numpy as np

# 读取数据
df = pd.read_csv(r"E:\大学\综合项目实践\processed_data.csv")
print(df.head())
# Min-Max 归一化函数
def min_max_normalization(column):
    return (column - column.min()) / (column.max() - column.min())

#对数据进行 Min-Max 归一化 
normalized_df = df.copy()
for column in df.columns[2:]:
    normalized_df[column] = min_max_normalization(normalized_df[column])

# 打印归一化后的数据前五行
print(normalized_df.head())

# 将归一化后的数据保存到新的 CSV 文件中
normalized_df.to_csv("Min-Max归一化pr.csv", index=False)
```

```python
# Z-Score 标准化函数
def z_score_normalization(column):
    return (column - column.mean()) / column.std()

# 对数据进行 Z-Score 标准化
normalized_df = df.copy()
for column in df.columns[2:]:
    normalized_df[column] = z_score_normalization(normalized_df[column])

# 打印归一化后的数据前五行
print(normalized_df.head())

# 将归一化后的数据保存到新的 CSV 文件中
normalized_df.to_csv("Z-Score标准化.csv", index=False)
```
